```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import patches
import os
import sys
```

```python
ROOT_DIR = os.path.abspath("../")
sys.path.append(ROOT_DIR)
```

```python
agg = os.path.join(ROOT_DIR, "dataset/processing_data/agg.csv")
agg_data = pd.read_csv(agg, encoding='utf-8', sep=',')
agg_data.head()
```

```python
image = plt.imread('../dataset/processing_data/clipped/1.tif')
plt.imshow(image)
```

```python
fig = plt.figure()

# add axes to the image
ax = fig.add_axes([0, 0, 1, 1])

# iterate over the image for object visualization
for _,row in agg_data[agg_data.image_name == "1.tif"].iterrows():
    
    xmin = row.minx
    xmax = row.maxx
    ymin = row.miny
    ymax = row.maxy
    
    width = xmax - xmin
    height = ymax - ymin
    
    print(ymin)
    
    ax.annotate('bird', xy=(xmax-10, ymin+10))
    edgecolor = 'r'
    
    # add bounding box to the image
    rect = patches.Rectangle((xmin, ymin), width, height, edgecolor=edgecolor, facecolor='none')
    #ax.add_patch(rect)
```
